```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import pandas as pd

# Carregar o conjunto de dados de ris
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Aplicar PCA para reduzir de 4 para 2 componentes principais
pca = PCA(n_components=2)
X_r = pca.fit_transform(X)

# Criar um DataFrame para facilitar a visualizao
df = pd.DataFrame(X_r, columns=['PC1', 'PC2'])
df['target'] = y
df['species'] = df['target'].apply(lambda i: target_names[i])

# Visualizar os dados transformados
plt.figure(figsize=(10, 7))
sns.scatterplot(x='PC1', y='PC2', hue='species', data=df, palette='Set1', s=100, alpha=0.7)
plt.title('PCA de ris: reduo para 2 componentes principais')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(loc='best', title='Espcies')
plt.grid(True)
plt.show()
```
