## 1. INTRODUÇÃO

Este notebook tem como objetivo analisar a relação entre a temperatura média diária e a quantidade de bicicletas alugadas, utilizando um modelo de Regressão Linear Simples. Os coeficientes do modelo serão determinados analiticamente por meio do método dos Mínimos Quadrados, implementado pela Equação Normal. O estudo utiliza um conjunto de dados real para ilustrar a aplicação prática da técnica em um problema de análise preditiva.

---

## 2. METODOLOGIA

O processo metodológico envolve a importação das bibliotecas, o carregamento e preparação dos dados a partir de um arquivo CSV, e o cálculo dos parâmetros do modelo de regressão.

### 2.1. Importação das bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Bibliotecas importadas.")
```

### 2.2. Carregamento e preparação dos dados

Os dados são lidos de um arquivo CSV. A coluna 'Temperatura_Media' é definida como a variável independente (X), e 'Bicicletas_Alugadas' como a variável dependente (y). A variável X é então preparada para o cálculo, adicionando-se um termo de interceptação.

```python
# Caminho para o arquivo de dados
caminho_csv = '../data/4.1 - dados_bicicletas.csv'

# Leitura dos dados
dados = pd.read_csv(caminho_csv)

# Preparação das variáveis X e y
X = dados['Temperatura_Media'].values.reshape(-1, 1)
y = dados['Bicicletas_Alugadas'].values

# Adição da coluna de 1s para o termo de interceptação (bias)
X_b = np.c_[np.ones((len(X), 1)), X]

print("Dados carregados e preparados para a modelagem.")
```

### 2.3. Cálculo dos coeficientes pela equação normal

Os parâmetros do modelo (intercepto e inclinação) são calculados de forma direta através da solução analítica da Equação Normal:

$ \hat{\theta} = (X^T X)^{-1} X^T y $

```python
# Cálculo dos coeficientes ótimos (theta)
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# Extração dos coeficientes individuais
intercepto = theta_best[0]
inclinação = theta_best[1]

print("Cálculo dos coeficientes concluído.")
```

## 3. RESULTADOS E ANÁLISE

Com os coeficientes do modelo definidos, os resultados são apresentados de duas formas: uma visualização gráfica do ajuste linear e a impressão dos valores numéricos dos coeficientes.

### 3.1. Visualização do ajuste linear

```python
# Geração de pontos para traçar a linha de regressão (de 0 a 30°C)
X_novo = np.array([[0], [30]])
X_novo_b = np.c_[np.ones((2, 1)), X_novo]
y_pred = X_novo_b.dot(theta_best)

# Visualização dos dados e do modelo de regressão
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.6, label='Dados reais')
plt.plot(X_novo, y_pred, color='red', linewidth=2, label='Linha de regressão')
plt.xlabel('Temperatura Média (°C)')
plt.ylabel('Bicicletas Alugadas')
plt.title('Regressão Linear: Efeito da Temperatura no Aluguel de Bicicletas')
plt.legend()
plt.grid(True)
plt.show()
```

### 3.2. Coeficientes do modelo

```python
# Impressão dos coeficientes do modelo
print(f"Intercepto (θ₀): {intercepto:.2f}")
print(f"Inclinação (θ₁): {inclinação:.2f}")
```
