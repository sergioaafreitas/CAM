```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Dados
anuncios_redes_sociais = np.array([3, 2, 4, 5, 3, 4, 5, 6, 4, 3])
anuncios_jornais = np.array([5, 3, 6, 2, 4, 5, 3, 2, 4, 6])
vendas = np.array([200, 150, 250, 300, 225, 275, 350, 400, 275, 225])

# Preparando os dados para a regressão
X = np.column_stack((anuncios_redes_sociais, anuncios_jornais))
y = vendas

# Criando o modelo de regressão linear múltipla
modelo = LinearRegression()
modelo.fit(X, y)

# Previsão
X_pred = np.array([[0, 0], [7, 7]])  # Exemplo de novos dados para previsão
y_pred = modelo.predict(X_pred)

# Plotando os resultados
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(anuncios_redes_sociais, anuncios_jornais, vendas, color='blue')
ax.set_xlabel('Anúncios Redes Sociais')
ax.set_ylabel('Anúncios Jornais')
ax.set_zlabel('Vendas do Dia')
ax.plot([0, 7], [0, 7], y_pred, color='red')  # Linha de regressão
plt.show()
```
