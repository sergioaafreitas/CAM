```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Dados
tamanho = np.array([120, 150, 80, 200, 100, 130, 160, 110, 180, 140])
quartos = np.array([3, 4, 2, 5, 3, 3, 4, 3, 4, 3])
idade = np.array([5, 7, 10, 3, 15, 4, 6, 8, 2, 12])
preco = np.array([300, 350, 200, 500, 250, 320, 360, 280, 480, 310])

# Preparando os dados para a regressão
X = np.column_stack((tamanho, quartos, idade))
y = preco

# Criando o modelo de regressão linear múltipla
modelo = LinearRegression()
modelo.fit(X, y)

# Coeficientes do modelo
print("Coeficientes: ", modelo.coef_)
print("Intercepto: ", modelo.intercept_)

# Previsão (exemplo)
previsao = modelo.predict([[150, 4, 5]])
print("Previsão de preço para casa de 150m², 4 quartos, 5 anos de idade: ", previsao)
```

- Importamos as bibliotecas necessárias para manipulação de dados e regressão linear.
- Definimos os dados de tamanho, quartos, idade e preço das casas.
- Preparamos os dados para a regressão, agrupando as variáveis independentes.
- Criamos e treinamos o modelo de regressão linear múltipla.
- Exibimos os coeficientes do modelo, que indicam a influência de cada variável independente no preço.
- Fazemos uma previsão de exemplo para uma casa com características específicas.
