```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.decomposition import PCA 

# Criando um dataset sintético com 3 classes em um espaço de alta dimensão
X, y = make_classification(n_samples=200, n_features=100, n_informative=5, n_classes=3, random_state=42)

# Aplicando PCA para reduzir o dataset de alta dimensão para 2 dimensões
pca = PCA(n_components=2)
X_r = pca.fit_transform(X)

# Criando um gráfico de dispersão dos dados originais nas duas primeiras dimensões
plt.figure(figsize=(14, 7))

# Gráfico de dispersão dos dados originais
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.title("Dados Originais nas Primeiras Duas Dimensões")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.grid(True)

# Gráfico de dispersão dos dados após a redução de dimensionalidade
plt.subplot(1, 2, 2)
plt.scatter(X_r[:, 0], X_r[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.title("Dados Após Redução de Dimensionalidade com PCA")
plt.xlabel("Primeira Componente Principal")
plt.ylabel("Segunda Componente Principal")
plt.grid(True)

plt.tight_layout()
plt.show()
```
