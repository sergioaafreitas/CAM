```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.datasets import load_iris
import pandas as pd

# Carregar o conjunto de dados de ris
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Aplicar LDA para reduzir de 4 para 2 componentes discriminantes
lda = LDA(n_components=2)
X_r2 = lda.fit_transform(X, y)

# Criar um DataFrame para facilitar a visualizao
df = pd.DataFrame(X_r2, columns=['LD1', 'LD2'])
df['target'] = y
df['species'] = df['target'].apply(lambda i: target_names[i])

# Visualizar os dados transformados
plt.figure(figsize=(10, 7))
sns.scatterplot(x='LD1', y='LD2', hue='species', data=df, palette='Set1', s=100, alpha=0.7)
plt.title('LDA de ris: reduo para 2 componentes discriminantes')
plt.xlabel('Discriminante Linear 1')
plt.ylabel('Discriminante Linear 2')
plt.legend(loc='best', title='Espcies')
plt.grid(True)
plt.show()

```
